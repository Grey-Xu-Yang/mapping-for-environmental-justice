```python
import pandas as np
import sqlite3
import pandas as pd
from pandas.io import sql
```

```python
con = sqlite3.connect("MEJ.db")
```

```python
sql_query = "SELECT * FROM census_tracts INNER JOIN exposure ON census_tracts.tract_id = exposure.tract_id INNER JOIN socioeconomic_factors ON census_tracts.tract_id = socioeconomic_factors.tract_id INNER JOIN environmental_effects ON census_tracts.tract_id = environmental_effects.tract_id INNER JOIN sensitive_population ON census_tracts.tract_id = sensitive_population.tract_id;"
```

```python
#pd.read_sql(sql_query, con, index_col = "Tract_ID")
```

```python
ct_df = pd.read_sql("SELECT * FROM census_tracts;", con, index_col = "Tract_ID")
print(ct_df.shape)
ct_df.head()
```

```python
exp_df = pd.read_sql("SELECT * FROM exposure;", con, index_col = "Tract_ID")
```

```python
print(exp_df.shape)
exp_df.head()
```

```python
sf_df = pd.read_sql("SELECT * FROM socioeconomic_factors;", con, index_col = "Tract_ID")
print(sf_df.shape)
sf_df.head()
```

```python
env_df = pd.read_sql("SELECT * FROM environmental_effects;", con, index_col = "Tract_ID")
print(env_df.shape)
env_df.head()
```

```python
sens_df = pd.read_sql("SELECT * FROM sensitive_population;", con, index_col = "Tract_ID")
print(sens_df.shape)
sens_df.head()
```

```python
#pd.concat([exp_df, sf_df, env_df, sens_df], join = "inner")
```

```python
CO_data = pd.merge(exp_df, sf_df, on = "Tract_ID").merge(env_df, on = "Tract_ID").merge(sens_df, on = "Tract_ID")
```

```python
CO_data.to_csv("CO_data.csv")
```
